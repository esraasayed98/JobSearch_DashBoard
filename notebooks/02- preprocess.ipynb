```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../data/Jobs_in_egypt.csv')
data.head()
```

# Remove duplicates


```python
job_column=data.pop('Job')
```

```python
data[data.duplicated()].index
```

```python
job_column.drop(data[data.duplicated()].index , inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
unique_data=pd.merge(data,job_column, left_on= data.index , right_on= job_column.index)
```

```python
unique_data.pop('key_0')
```

```python
unique_data.reset_index(drop=True, inplace=True)
```

```python
unique_data.info()
```

# Remove confidential company

```python
unique_data=unique_data[unique_data['company'] != 'Confidential Company -\xa0']
unique_data
```

```python
unique_data.reset_index(drop=True, inplace=True)
unique_data
```

```python
unique_data.to_csv('../data/unique data.csv', sep=',', index=False)
```
